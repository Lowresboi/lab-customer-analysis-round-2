```python
# Importing all the neccesary and then mount my drive:
import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive')
```

```python
# Read the file for the lab, I will call it file_mkt now on:
file_mkt = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Labs 3/lab-customer-analysis-round-2/files_for_lab/csv_files/marketing_customer_analysis.csv')
file_mkt
```

```python
# Showing the shape of the file:
file_mkt.shape
```

```python
# Standarize the data (I only put it on upper):
file_mkt.columns = file_mkt.columns.str.upper()
file_mkt
```

```python
# Showing the tipes of data:
file_mkt.dtypes

# Numerical and categorical (I used this for this time):
numerical = file_mkt.select_dtypes(include=[np.number]).columns
print(numerical)

categorical = file_mkt.select_dtypes(include=[np.object]).columns
print(categorical)
```

```python
# Checking NaN in the file:
nan = file_mkt.isnull().sum()
print(nan)

# Droping NaN:
file_mkt = file_mkt.dropna()
print(file_mkt)
```

```python
# Datetime formating:
file_mkt['EFFECTIVE TO DATE'] = pd.to_datetime(file_mkt['EFFECTIVE TO DATE'])
file_mkt['month'] = file_mkt['EFFECTIVE TO DATE'].dt.month

# Display (used a little bit of help here):
first_quarter_data = file_mkt[file_mkt['month'].isin([1, 2, 3])]
print(first_quarter_data.head())
```
